```python
import pandas as pd
```

```python
unemployment = pd.read_csv("CAUR.csv")
```

```python
newdate = pd.date_range('1976-01', '2017-01', freq='M')
newdate
```

```python
unemployment.ix[:,"date"] = newdate
```

```python
unemployment.head()
```

```python
unemployment.dtypes
```

```python
houseprice = pd.read_csv("CA_sold_t.csv")
```

```python
houseprice.ix[:,"Median"] = houseprice.median(axis=1)
houseprice.head()
```

```python
newdate = pd.date_range('1996-04', '2016-07', freq='M')
houseprice.ix[:,"date"] = newdate
```

```python
houseprice
```

```python
UR_houseprice = pd.merge(houseprice, unemployment, on='date')
UR_houseprice = UR_houseprice.set_index('date')
UR_houseprice.head()
```

```python
from plotly import __version__, tools
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.plotly as py
import plotly.graph_objs as go
init_notebook_mode(connected=True)

houseprice = go.Scatter(
    x= UR_houseprice.index,
    y= UR_houseprice.Median,
    name = "Median House Price",
    yaxis =  "y1"
)

unemployment = go.Scatter(
    x= UR_houseprice.index,
    y= UR_houseprice.Unemployment,
    name = "Unemployment Rate",
    yaxis =  "y2"
)

data = [houseprice,unemployment]
layout = go.Layout(
    title='Sold Price and Unemployment Rate from 1996 to 2016 in Califorina',
    yaxis=dict(
        title='Median House Price',
        zeroline = False
    ),
    yaxis2=dict(
        title='Unemployment Rate (%)',
        overlaying='y',
        side='right',
        zeroline = False
    ),
    legend = dict(
        x =  1.04761904762, 
        y =  0.966666666667, 
    )
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
