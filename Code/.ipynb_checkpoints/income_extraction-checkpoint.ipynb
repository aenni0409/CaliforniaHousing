```python
import pandas as pd
```

```python
# data comes from U.S. census bureau
# only extracted the some column related to income for all household
income = pd.read_excel("ACS_15_5YR_S1901.xls").ix[9:,[0,2,3,5]]
```

```python
# rename the columns to make data manipulation easier
income.rename(columns={
        "Unnamed: 2" : "household_type",
        "Unnamed: 3" : "kind",
        "Unnamed: 5" : "total_income"
    },inplace = True)
income.head()
```

```python
income = income[(income.household_type == "Households") & (income.kind == "Estimate")]
```

```python
income = income.reset_index(drop = True)
```

```python
county = income.ix[:,0]
county_split = county.str.split(",")
# do string method to extract the county names without county and California
newcounty = [county_split[i][0] for i in xrange(len(county_split))]
newcounty = [newcounty[i].replace("County","").strip() for i in xrange(len(newcounty))]
```

```python
income["county"] = newcounty
```

```python
# turn total_income to int
income["total_income"] = income["total_income"].str.replace(",","")
income[["total_income"]] = income[["total_income"]].astype(int)
income = income[[1,2,3,4]]
```

```python
# output the data
income.to_csv('income.csv')
```
